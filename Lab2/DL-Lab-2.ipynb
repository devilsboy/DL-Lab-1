**Importing necessary libraries**

```python
import pandas as pd
import statistics
import numpy as np
```

# Step1

**Reading Data**

```python
data = pd.read_excel(r'Online Retail.xlsx')
```

**Indexing**

```python
subset = data.iloc[0:536544, 0:4]
print(subset)
```

**Importing only stockCode and unitPrice**

```python
price = data["UnitPrice"]#Read Price
stockCode = data["StockCode"]#Read StockCode
```

**Statistical analysis**

```python
print("Min Price:",min(price))#Min Price

print("Max Price:",max(price))#Max Price

print("Median:",statistics.median(price))#Median

print("Mean:",statistics.mean(price))#Mean

print("Count:",len(price))#Count

print("Sum:",sum(price))#Sum

print("Standard Deviation:",statistics.stdev(price))#Standard Deviation

print("Variance:",statistics.variance(price))#Variance
```

**Creating Panda Series with random integers**

```python
a = pd.Series(np.random.randint(0,100,size=(100)))
b = pd.Series(np.random.randint(0,100,size=(100)))
```

**Operations on the series**

```python
add = a+b
sub = a-b
mult = a*b
div = a/b
```

**Creating a dataframe from this series**

```python
data = {'number1':a,'number2':b,'add':add,'sub':sub,'mult':mult,'div':div}
df = pd.DataFrame(data)
df
```
