```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Advertising.csv")
data.head()
```

```python
data.isnull()
```

```python
sales_type = data["Sales"].tolist()
print(sales_type)
```

```python
for i in range(len(sales_type)):
    if sales_type[i] >= 15:
        sales_type[i] = "high"
    elif sales_type[i] < 15 and sales_type[i] >= 5:
        sales_type[i] = "medium"
    else:
        sales_type[i] = "low"
```

```python
print(sales_type)
```

```python
data["sales_type"] = sales_type
data.head()
```

```python
grouped = data.groupby("sales_type")
high = grouped.get_group("high")
```

```python
print("TV SUM:",sum(high["TV"]))
print("TV MEAN:",np.mean(high["TV"]))
print("RADIO SUM:",sum(high["Radio"]))
print("RADIO MEAN:",np.mean(high["Radio"]))
print("NEWSPAPER SUM:",sum(high["Newspaper"]))
print("NEWSPAPER MEAN:",np.mean(high["Newspaper"]))
```

```python
med = grouped.get_group("medium")
print("TV SUM:",sum(med["TV"]))
print("TV MEAN:",np.mean(med["TV"]))
print("RADIO SUM:",sum(med["Radio"]))
print("RADIO MEAN:",np.mean(med["Radio"]))
print("NEWSPAPER SUM:",sum(med["Newspaper"]))
print("NEWSPAPER MEAN:",np.mean(med["Newspaper"]))
```

```python
low = grouped.get_group("low")
print("TV SUM:",sum(low["TV"]))
print("TV MEAN:",np.mean(low["TV"]))
print("RADIO SUM:",sum(low["Radio"]))
print("RADIO MEAN:",np.mean(low["Radio"]))
print("NEWSPAPER SUM:",sum(low["Newspaper"]))
print("NEWSPAPER MEAN:",np.mean(low["Newspaper"]))
```

```python
for i in range(1,len(data["TV"])):
    if i % 6 == 0 and i > 0:
        data["TV"][i] = 'NaN'
    elif i % 10 == 0 and i > 0:
        data["Newspaper"][i] = 'NaN'
print(data)
```

```python
bool_series = pd.notnull(data["TV"]+data["Newspaper"])
data[bool_series]
```

```python
data.fillna(0)
```

```python
data = data.fillna(data.mean())
data
```

```python
normal = preprocessing.normalize([data["TV"]])
norm = normal[0]
data["TV"] = norm
normal = preprocessing.normalize([data["Radio"]])
norm = normal[0]
data["Radio"] = norm
normal = preprocessing.normalize([data["Newspaper"]])
norm = normal[0]
data["Newspaper"] = norm
```

```python
x = data.iloc[:,1:4]
y = data.iloc[:,4:5]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.40,random_state = 0)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train) 
```

```python
y_pred = regr.predict(x_test)
print("Mean square Error =",mean_squared_error(y_pred,y_test))
```

```python
print("Accuracy on testing=",regr.score(x_test, y_test))
```
